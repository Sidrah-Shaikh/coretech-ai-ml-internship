1 What is Artificial Intelligence?
Answer:
AI is the ability of computers to perform tasks that normally require human intelligence.

2 What is Machine Learning?
Answer:
Machine Learning is a branch of AI that allows computers to learn from data and improve automatically.

3 What is Deep Learning?
Answer:
Deep Learning is a subset of Machine Learning that uses neural networks to solve complex problems.

4 Difference between AI, ML and DL
Answer:
AI: Makes machines intelligent.
ML: Helps machines learn from data.
DL: Uses deep neural networks for advanced learning.
