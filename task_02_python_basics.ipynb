# Example 1: String
name = "Sidra"
print("Name:", name)

# Example 2: Integer
age = 20
print("Age:", age)

# Example 3: Float
height = 5.3
print("Height:", height)

# Example 4: Boolean
is_student = True
print("Student:", is_student)

# Example 5: List
skills = ["Python", "Graphic Design", "Video Editing"]
print("Skills:", skills)

# Example 6: Tuple
colors = ("Red", "Green", "Blue")
print("Colors:", colors)

# Example 7: Dictionary
student = {
    "Name": "Sidra",
    "Age": 20,
    "City": "Mehrabpur"
}
print("Student Info:", student)

# Example 8: Conditional Statement
if age >= 18:
    print("Adult")
else:
    print("Minor")

# Example 9: For Loop
print("For Loop:")
for i in range(1, 6):
    print(i)

# Example 10: While Loop
print("While Loop:")
count = 1
while count <= 5:
    print(count)
    count += 1

# Example 11: Function
def greet(name):
    return "Hello " + name

print(greet("Sidra"))

# Example 12: List Comprehension
squares = [x * x for x in range(1, 6)]
print("Squares:", squares)

# Example 13: NumPy Basics
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print("NumPy Array:", arr)
print("Mean:", arr.mean())

# Example 14: Pandas Basics
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "Ahmed"],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)
print("\nPandas DataFrame:")
print(df)
