```python
import pandas as pd

data = {
    'StudentID':[201,202,203,204,204,205],
    'Age':[18,20,19,21,21,22],
    'Gender':['Female','Male','Female','Male','Male','Female'],
    'Course':['CS','IT','SE','CS','CS','AI'],
    'Marks':[85,None,78,92,92,88],
    'Passed':['Yes','Yes','No','Yes','Yes','Yes']
}

df = pd.DataFrame(data)

df.to_csv('student_data.csv', index=False)

print("student_data.csv created successfully")
```

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, RobustScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif

df = pd.read_csv("student_data.csv")

# Fill missing values
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

# Remove duplicates
df.drop_duplicates(inplace=True)

# Label Encoding
le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])
df["Passed"] = le.fit_transform(df["Passed"])

# One Hot Encoding
df = pd.get_dummies(df, columns=["Course"])

# Scaling
scaler = RobustScaler()
df[["Age","Marks"]] = scaler.fit_transform(df[["Age","Marks"]])

# Feature Selection
X = df.drop("Passed", axis=1)
y = df["Passed"]

selector = SelectKBest(score_func=mutual_info_classif, k=3)
selector.fit(X, y)

print("Selected Features:")
print(X.columns[selector.get_support()])

# Save cleaned data
df.to_csv("cleaned_student_data.csv", index=False)

# Visualization
before = 6
after = len(df)

plt.figure(figsize=(6,4))
plt.bar(["Original Data","Cleaned Data"], [before, after])
plt.title("Student Dataset Cleaning")
plt.savefig("student_chart.png")
plt.show()

print("Task Completed Successfully")
```
